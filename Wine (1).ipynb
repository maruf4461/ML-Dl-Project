## Developing Random forest classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

# Label encode the 'quality' column
label_encoder = LabelEncoder()
df_merged['quality_encoded'] = label_encoder.fit_transform(df_merged['quality'])

# Split the data into features and target
X = df_merged.drop(['quality', 'quality_encoded'], axis=1)
y = df_merged['quality_encoded']

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled.shape, X_test_scaled.shape, y_train.shape, y_test.shape
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
# Redefine and retrain the Random Forest model
rf_classifier = RandomForestClassifier(random_state=0)
rf_classifier.fit(X_train_scaled, y_train)

# Predict probabilities for each class
y_pred_proba_rf = rf_classifier.predict_proba(X_test_scaled)

# Calculate the macro-average ROC AUC score across all classes
roc_auc_rf = roc_auc_score(y_test, y_pred_proba_rf, multi_class="ovr", average="macro")

# Recalculate predictions for confusion matrix and classification report
y_pred_rf = rf_classifier.predict(X_test_scaled)
rf_class_report = classification_report(y_test, y_pred_rf)
rf_conf_matrix = confusion_matrix(y_test, y_pred_rf)

print('ROC AUC:',roc_auc_rf)
print('Classification report: ')
print(rf_class_report)
print('Confusion matrix: ')
print(rf_conf_matrix)
```

## Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

# Train the Gradient Boosting Classifier
gbm_classifier = GradientBoostingClassifier(random_state=0)
gbm_classifier.fit(X_train_scaled, y_train)

# Predict probabilities for each class
y_pred_proba_gbm = gbm_classifier.predict_proba(X_test_scaled)

# Calculate the macro-average ROC AUC score across all classes
roc_auc_gbm = roc_auc_score(y_test, y_pred_proba_gbm, multi_class="ovr", average="macro")

# Predict labels for confusion matrix and classification report
y_pred_gbm = gbm_classifier.predict(X_test_scaled)
gbm_class_report = classification_report(y_test, y_pred_gbm)
gbm_conf_matrix = confusion_matrix(y_test, y_pred_gbm)

print('ROC AUC',roc_auc_gbm)
print('Classification report:')
print(gbm_class_report)
print('Confusion matrix:')
print(gbm_conf_matrix)
```

## Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

# Train the Support Vector Machine
svm_classifier = SVC(probability=True, random_state=0)
svm_classifier.fit(X_train_scaled, y_train)

# Predict probabilities for each class
y_pred_proba_svm = svm_classifier.predict_proba(X_test_scaled)

# Calculate the macro-average ROC AUC score across all classes
roc_auc_svm = roc_auc_score(y_test, y_pred_proba_svm, multi_class="ovr", average="macro")

# Predict labels for confusion matrix and classification report
y_pred_svm = svm_classifier.predict(X_test_scaled)
svm_class_report = classification_report(y_test, y_pred_svm)
svm_conf_matrix = confusion_matrix(y_test, y_pred_svm)

print(roc_auc_svm)
print(svm_class_report)
print(svm_conf_matrix)
```

## SVM using SMOTE Balancing class

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as IMBPipeline
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assuming X_train_scaled, y_train are defined from your dataset as shown earlier

# Create an SVM model with SMOTE for handling class imbalance
svm_smote_pipeline = IMBPipeline([
    ('smote', SMOTE(random_state=0)),
    ('svm', SVC(probability=True, random_state=0))
])

# Train the SVM with SMOTE
svm_smote_pipeline.fit(X_train_scaled, y_train)

# Predict and evaluate
y_pred_svm_smote = svm_smote_pipeline.predict(X_test_scaled)
print(classification_report(y_test, y_pred_svm_smote))
print(confusion_matrix(y_test, y_pred_svm_smote))
```

## SVM using class weight to handle class imbalance

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Setup for class weights: 'balanced' automatically adjusts weights inversely proportional to class frequencies
svm_classifier = SVC(class_weight='balanced', probability=True, random_state=0)

# Train the SVM with class weight adjustment
svm_classifier.fit(X_train_scaled, y_train)

# Predict and evaluate
y_pred_svm_weighted = svm_classifier.predict(X_test_scaled)
print(classification_report(y_test, y_pred_svm_weighted))
print(confusion_matrix(y_test, y_pred_svm_weighted))
```

## Neural Network by improving learning rate activation, epochs and batch size and hidden layers

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as IMBPipeline
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Assuming your data is already loaded and split into X_train, X_test, y_train, y_test

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the MLPClassifier with increased epochs
nn_improved_epochs = MLPClassifier(
    hidden_layer_sizes=(100,),
    activation='relu',
    solver='adam',
    alpha=0.005,
    batch_size=200,
    learning_rate_init=0.002,
    max_iter=100,  # Increased number of epochs
    random_state=0,
    verbose=True
)

# Create a pipeline with SMOTE and the improved MLP classifier
nn_smote_pipeline = IMBPipeline([
    ('smote', SMOTE(random_state=0)),
    ('nn', nn_improved_epochs)
])

# Train the Neural Network with SMOTE
nn_smote_pipeline.fit(X_train_scaled, y_train)

# Predict and evaluate the model with SMOTE
y_pred_nn_smote = nn_smote_pipeline.predict(X_test_scaled)
print(classification_report(y_test, y_pred_nn_smote))
print(confusion_matrix(y_test, y_pred_nn_smote))
```
